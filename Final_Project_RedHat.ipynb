```
#merge data
people_train = pd.merge(train, people1, on='people_id')
people_test = pd.merge(test, people1, on='people_id')
```

```
#group 17304 gives zero outcome, so delete from table since it's not neccessary
people_train.loc[people_train['group_1'] == 'group 17304', ['outcome']]
```

```
#dropping group 17304 and filling in NA's with 0.
people_train1 = people_train.set_index("group_1")
people_train1 = people_train1.drop('group 17304', axis = 0)
people_train1 = people_train.fillna(0, axis=0)
people_train1
```

```
#remove group 17304 from people_test and replacing NA's with 0.
people_test1 = people_test.set_index("group_1")
people_test1 = people_test1.drop('group 17304', axis = 0)
people_test1 = people_test1.fillna(0, axis=0)
people_test1
```

```
'''change the group_1 values from string to integer, by taking out the "group"
part of the value and keeping the numbers, and then converting string to int.'''
people_train1.group_1 = people_train1.group_1.str.lstrip('group')
for i in people_train1.group_1:
  i = int(i)
  
people_train1.group_1
```

```
#naive bayes model
import numpy as np
from sklearn.model_selection import train_test_split

X = people_train1['group_1'].values #feature
y = people_train1['outcome'].values #dependent variable

train_X, test_X, train_y, test_y = train_test_split(X, y, 
                                                    train_size=0.7,
                                                    test_size=0.3,
                                                    random_state=100)
train_X = train_X.reshape(-1, 1)
train_y = train_y.reshape(-1, 1)
test_X = test_X.reshape(-1, 1)

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(train_X, train_y)
accuracy_result = clf.score(test_X, test_y)
print('naive bayes model accuracy:', accuracy_result)
```

```
#K-Nearest Neighbor Model
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics

classifier = KNeighborsClassifier()
classifier.fit(train_X, train_y)

prediction_accuracy_test = classifier.predict(test_X)

accuracy_test = metrics.accuracy_score(test_y, prediction_accuracy_test)

print('Accuracy of test dataset:', accuracy_test)
```

```
#support vector machines model, but took too long to run so was stopped.
from sklearn.svm import SVC
clf = SVC(gamma='auto')
clf.fit(train_X, train_y)
accuracy_result1 = clf.score(test_X, test_y)
print('support vector machines model accuracy:', accuracy_result1)
```

```
#random forest model
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators=100, oob_score = True, n_jobs = -1, 
                             random_state = 0, max_features = None, 
                             min_samples_leaf = 30)
clf.fit(train_X, train_y)

accuracy_score1 = clf.score(test_X, test_y)

print('random forest model accuracy:', accuracy_score1)
```
